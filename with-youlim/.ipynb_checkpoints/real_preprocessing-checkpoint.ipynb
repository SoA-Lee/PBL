```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import re
import json
from konlpy.tag import Okt
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
from tensorflow.python.keras.preprocessing.text import Tokenizer
%matplotlib inline
```

```python
DATA_IN_PATH = './data_in/'
print("파일 크기 : ")
for file in os.listdir(DATA_IN_PATH):
    if 'csv' in file :
        print(file.ljust(30) + str(round(os.path.getsize(DATA_IN_PATH + file) / 1000000, 2)) + 'MB')
```

```python
train_data = pd.read_csv(DATA_IN_PATH+'real_train_data.csv', names=['sentence', 'label'])
test_data = pd.read_csv(DATA_IN_PATH+'real_test_data.csv',names=['sentence', 'label'])
```

```python
#데이터 정제/ 전처리 함수
def preprocessing(sentence, okt, remove_stopwords=False, stop_words=[]):
    #remove_stopwords: 불용어 제거 여부 선택
    #review_text ex)아 더빙 진짜 짜증나네요 목소리
    word_sentence= okt.morphs(sentence, stem=True) #형태소 단위로 분리
    print(word_sentence) #ex)['아', '더빙', '진짜', '짜증나다', '목소리']
    if remove_stopwords:
        preprocessed_word_sentence = [token for token in word_sentence if not token in stop_words]
        #preprocessed_word_rewiew ex)['더빙', '진짜', '짜증나다', '목소리']
    return preprocessed_word_sentence
```

```python
#preprocessing함수 train_data
stop_words = set(['은', '는', '이', '가', '하', '아', '것', '들', '의', '있', '되', '수', '보', '주', '등', '한'])
okt = Okt()
train_clean_sentence = []

for sentence in train_data['sentence']:
    #빈데이터에 의한 중단 방지
    if type(sentence) == str:
        train_clean_sentence.append(preprocessing(sentence, okt, remove_stopwords=True, stop_words=stop_words))
    else:
        train_clean_sentence.append([])
        
complete_preprocessing_train_df = pd.DataFrame({'sentence': train_clean_sentence, 'label': train_data['label']})
```

```python
#preprocessing함수 test_data
test_clean_sentence = []

for sentence in test_data['sentence']:
    #빈데이터에 의한 중단 방지
    if type(sentence) == str:
        test_clean_sentence.append(preprocessing(sentence, okt, remove_stopwords=True, stop_words=stop_words))
    else:
        test_clean_sentence.append([])
        
complete_preprocessing_testonlysentence_df = pd.DataFrame({'sentence': test_clean_sentence})
complete_preprocessing_test_onlylabel_df = pd.DataFrame({'label': test_data['label']})
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_clean_sentence) ####only train data만 단어사전에 
train_sequences = tokenizer.texts_to_sequences(train_clean_sentence)
test_sequences = tokenizer.texts_to_sequences(test_clean_sentence)
word_vocab = tokenizer.word_index
# word_vocab["<PAD>"] = 0

MAX_SEQUENCE_LENGTH = 36

train_inputs = pad_sequences(train_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post') #train data벡터화
train_labels = np.array(train_data['label'])
test_inputs = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post') #test data 벡터화
test_labels = np.array(test_data['label'])
```
